```python
import pandas as pd
```

```python
us11 = pd.read_csv('11_national.csv')
us_previous = pd.to_numeric(us11[1:]['HD01_VD01'][1])
us16 = pd.read_csv('16_national.csv')
us_latest = pd.to_numeric(us16[1:]['HD01_VD01'][1]) 
nj11 = pd.read_csv('11_state.csv')
nj_previous =pd.to_numeric(nj11[1:]['HD01_VD01'][1])
nj16 = pd.read_csv('16_state.csv')
nj_latest = pd.to_numeric(nj16[1:]['HD01_VD01'][1]) 
```

```python
nj_latest
```

```python
muni16 = pd.read_csv('16_muni.csv', 
                     na_values=['-', '**', '***', '*****'], 
                     usecols = [ 'GEO.id', 'GEO.id2','GEOdisplaylabel_long', 
                           'pop_latest','pop_latest_moe' ],
                     header=None,
                     skiprows=2,
                    names=[ 'GEO.id', 'GEO.id2','GEOdisplaylabel_long', 
                           'pop_latest','pop_latest_moe'  ])
cty16 = pd.read_csv('16_county.csv',
                    na_values=['-', '**','***', '*****'],
                   usecols = [ 'GEO.id', 'GEO.id2','GEOdisplaylabel_long', 
                           'pop_latest','pop_latest_moe'  ],
                    header=None,
                    skiprows=2,
                   names=[ 'GEO.id', 'GEO.id2','GEOdisplaylabel_long', 
                           'pop_latest','pop_latest_moe'  ])

pop16 = muni16.append(cty16).reset_index().drop(labels='index', axis=1)
```

```python
muni11 = pd.read_csv('11_muni.csv', 
                     na_values=['-', '**', '***', '*****'], 
                     usecols = [ 'GEO.id', 'GEO.id2','GEOdisplaylabel_long', 
                           'pop_prev','pop_prev_moe' ],
                     header=None,
                     skiprows=2,
                    names=[ 'GEO.id', 'GEO.id2','GEOdisplaylabel_long', 
                           'pop_prev','pop_prev_moe' ])
cty11 = pd.read_csv('11_county.csv',
                    na_values=['-', '**','***', '*****'],
                   usecols = [ 'GEO.id', 'GEO.id2','GEOdisplaylabel_long', 
                           'pop_prev','pop_prev_moe' ],
                    header=None,
                    skiprows=2,
                   names=[ 'GEO.id', 'GEO.id2','GEOdisplaylabel_long', 
                           'pop_prev','pop_prev_moe' ])

pop11 = muni11.append(cty11).reset_index().drop(labels='index', axis=1)
```

```python
pop16.head()
```

```python
pop16['GEOdisplaylabel_long'] = pop16['GEOdisplaylabel_long'].replace(to_replace=', New Jersey', value='', regex=True)
pop11['GEOdisplaylabel_long'] = pop11['GEOdisplaylabel_long'].replace(to_replace=', New Jersey', value='', regex=True)
```

```python
pop_cols = pd.DataFrame(pop16.GEOdisplaylabel_long.str.split(',',1).tolist(), columns = ['GEOdisplaylabel', 'County'])
pop16 = pd.concat([pop16, pop_cols ], axis=1)
pop16['County'] = pop16['County'].replace(to_replace='County', value='', regex=True)
pop16.head()
```

```python
pop = pd.merge(pop16, pop11, on='GEO.id2', how='left')
```

```python
pop.head()
```

```python
pop_simp = pop[[u'GEO.id_x', u'GEO.id2', u'GEOdisplaylabel', u'GEOdisplaylabel_long_x',u'County',
       u'pop_latest', u'pop_latest_moe', 
        u'pop_prev',
       u'pop_prev_moe']].rename(columns={'GEO.id_x':'GEO.id', 
                                                'GEOdisplaylabel_long_x':'GEOdisplaylabel_long'})
```

```python
pop_simp.head()
```

```python
pop_simp = pop_simp[pop_simp['GEOdisplaylabel'] != 'County subdivisions not defined']
```

```python
pop_simp.head()
```

```python
pop_simp['pop_moe_latest_per'] = (pop_simp['pop_latest_moe']/pop_simp['pop_latest'])*100 
pop_simp['pop_moe_prev_per'] = (pop_simp['pop_prev_moe']/pop_simp['pop_prev'])*100 
```

```python
pop_simp.head()
```

```python
#Add NJ and Nationwide
pop_simp['nj_pop_latest'] = nj_latest
pop_simp['nj_pop_prev'] = nj_previous
pop_simp['usa_pop_latest'] = us_latest
pop_simp['usa_pop_prev'] = us_previous
pop_simp['pop_per_change'] = ((pop_simp['pop_latest'] - pop_simp['pop_prev']) / pop_simp['pop_prev']) * 100
pop_simp['nj_pop_change'] = ((pop_simp['nj_pop_latest'] - pop_simp['nj_pop_prev']) / pop_simp['nj_pop_prev']) * 100
pop_simp['us_pop_change'] = ((pop_simp['usa_pop_latest'] - pop_simp['usa_pop_prev']) / pop_simp['usa_pop_prev']) * 100
```
